# Saving df to a db

```python
import os
import csv
import mysql.connector
```

```python
conn = mysql.connector.connect(host='localhost',user='newuser',passwd='',use_pure=True,database='study_db')
```

```python
curs = conn.cursor()
```

```python
curs.execute('''
CREATE TABLE PACKAGE(
CGPA float,
pack float)''')
```

```python
file = open('/Users/joyalchully/Desktop/Summer 2022/Campusx/package.csv')
reader = csv.reader(file,delimiter=',')
```

```python
for num,row in enumerate(reader):
    if num==0:
        pass
    else:
        print(num,row[0])
        cgpa = row[0]
        package = row[1]
        curs.execute('''
        insert into package(cgpa,pack)
        values({},{})'''.format(cgpa,package))
        conn.commit()
    
```
